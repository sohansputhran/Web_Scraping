```python
import pandas as pd

from bs4 import BeautifulSoup
import requests
import urllib.request
```

```python
authors = []
quotes = []
```

```python
def scrape_website(page_number):
    
    URL = 'https://www.goodreads.com/quotes/tag/inspirational?page=' + str(page_number)
    webpage = requests.get(URL)
    soup = BeautifulSoup(webpage.text, 'html.parser')
    quoteText = soup.find_all('div', attrs= { 'class': 'quoteText'})

    for i in quoteText:
        quote = i.text.strip().split('\n')[0]
        author = i.find('span', attrs = { 'class': 'authorOrTitle'}).text.strip()
        quotes.append(quote)
        authors.append(author)
```

```python
for page_num in range(0,10):
    scrape_website(page_num)
```

```python
combined_list = []
for index in range(len(quotes)):
    combined_list.append(quotes[index] + " - " + authors[index])
```
